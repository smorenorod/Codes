```python
variables=nba_data.columns[1:20]
variables

r,c=5,4
pos = [(x+1, y+1) for x in range(r) for y in range(c)]

fig = make_subplots(rows=r, cols=c)

for col,s in zip(variables,pos):

    fig.add_trace(go.Box(y=nba_data[col],name=col),row=s[0], col=s[1])

fig.update_layout(height=1000, width=
                  900, title_text="Boxplots By Feature")
fig.show()
```

```python
#variables with outliers

outlier_var=['pts', 'fgm', 'fga', 'fg', '3p_made', '3pa','3p','ftm', 'fta', 'ft', 'oreb', 'dreb', 'reb', 'ast', 'stl', 'blk', 'tov']
non_outl_var=['gp', 'min']
nba_data[outlier_var]
```

```python
#feature/target split

X = nba_data.drop(['name','target_5yrs'], axis=1)
#nba_data[nba_data.columns[~nba_data.columns.isin(['name','target_5yrs'])]]

Y=nba_data["target_5yrs"]
```

```python
#train/test split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#check PCA number of dimensions#

pca = PCA()
pca.fit(X_train)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
plt.plot(cumsum)
plt.axvline(x=d, color='k', linestyle='--')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')

print("Number of components:", d)
```

```python
#check Y distribution

100*nba_data.groupby(by= 'target_5yrs')["target_5yrs"].count()/len(nba_data)
```

```python
#create a simple pipeline for all columns for logreg
pipe = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=4)),
			('clf', LogisticRegression(random_state=42))])

pipe.fit(X_train, y_train)



y_pred = pipe.predict(X_test)

print('ROCAUC Score: {:.2f}'.format(metrics.roc_auc_score(y_test, y_pred)))
print('Accuracy: {:.2f}'.format(metrics.accuracy_score(y_test, y_pred)))
print('F1 Score: {:.2f}'.format(metrics.f1_score(y_test, y_pred)))
```

```python
#create a simple pipeline for all columns for svm clasifier
pipe = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=4)),
			('clf', svm.SVC(random_state=42))])

pipe.fit(X_train, y_train)



y_pred = pipe.predict(X_test)

print('ROCAUC Score: {:.2f}'.format(metrics.roc_auc_score(y_test, y_pred)))
print('Accuracy: {:.2f}'.format(metrics.accuracy_score(y_test, y_pred)))
print('F1 Score: {:.2f}'.format(metrics.f1_score(y_test, y_pred)))
```

```python
#create a simple pipeline for all columns for DecisionTreeClassifier
pipe = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=4)),
			('clf', DecisionTreeClassifier(random_state=42))])

pipe.fit(X_train, y_train)



y_pred = pipe.predict(X_test)

print('ROCAUC Score: {:.2f}'.format(metrics.roc_auc_score(y_test, y_pred)))
print('Accuracy: {:.2f}'.format(metrics.accuracy_score(y_test, y_pred)))
print('F1 Score: {:.2f}'.format(metrics.f1_score(y_test, y_pred)))
```

```python
#create a simple pipeline for all columns for NN Classifier
pipe = Pipeline([('scl', StandardScaler()),
			('pca', PCA(n_components=4)),
			('clf', MLPClassifier(random_state=42,max_iter=600))])

pipe.fit(X_train, y_train)



y_pred = pipe.predict(X_test)

print('ROCAUC Score: {:.2f}'.format(metrics.roc_auc_score(y_test, y_pred)))
print('Accuracy: {:.2f}'.format(metrics.accuracy_score(y_test, y_pred)))
print('F1 Score: {:.2f}'.format(metrics.f1_score(y_test, y_pred)))
```

```python
parameter_space = {
    'clf__hidden_layer_sizes': [(60+5*i, ) for i in range(10)],
    'clf__activation': ['logistic', 'tanh', 'relu'],
    'clf__solver': ['sgd', 'adam'],
    #'clf__alpha': 10.0 ** -np.arange(1, 10
}



grid = GridSearchCV(pipe, parameter_space, n_jobs= -1, cv=5,scoring ='f1')

grid.fit(X_train, y_train)

print('Grid best parameter: ',grid.best_params_)
print('Grid best f1: ', grid.best_score_)
```

```python
# create the preprocessing pipelines for both outlier and non outlier data.

outl_transformer = Pipeline(steps=[
    ('pow_transf',  PowerTransformer(method='yeo-johnson',standardize=True)), #using yeo-johnson since 'box-cox' admint only  strictly positive and there are 0 present
    #('scl', StandardScaler()), #no scaller needed since box_cox is doing the standarization
    #('pca', PCA(n_components=4))
    #('imputer', SimpleImputer(strategy='constant', fill_value='missing')), # no imputer needed since non missing values
    #('onehot', OneHotEncoder(handle_unknown='ignore'))  # no one hot needed since there is not categorical data present
    ])


non_outl_transformer = Pipeline(steps=[
    ('scl', StandardScaler()),
    #('pca', PCA(n_components=4)),
    #('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    #('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])


preprocessor = ColumnTransformer(
    transformers=[
        ('outliar', outl_transformer, outlier_var),
        ('nonoutliar', non_outl_transformer, non_outl_var)])


# Append classifier to preprocessing pipeline.
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('pca', PCA(n_components=4)), #adding PCA to the total of features
                      ('classifier', LogisticRegression(solver='liblinear'))]) #‘liblinear’ and ‘saga’ handle L1 and L2 penalty
```

```python
clf.fit(X_train, y_train)


y_pred = clf.predict(X_test)

print('ROCAUC Score: {:.2f}'.format(metrics.roc_auc_score(y_test, y_pred)))
print('Accuracy: {:.2f}'.format(metrics.accuracy_score(y_test, y_pred)))
print('F1 Score: {:.2f}'.format(metrics.f1_score(y_test, y_pred)))
```

```python

parameter_space = {
    'classifier__C': np.logspace(-3, 3, 7),
    'classifier__penalty': ["l1", "l2"],# l1 lasso l2 ridge
    'pca__n_components': np.arange(1, 17, 2)  }

grid = GridSearchCV(clf, parameter_space, n_jobs= -1, cv=5,scoring =["accuracy","roc_auc","f1"], refit='roc_auc')
grid.fit(X_train, y_train)

print('Grid best parameter: ',grid.best_params_)
print('Grid best f1: ', grid.best_score_)

grid_pred = grid.predict(X_test)
```

```python


print(grid.cv_results_['mean_test_accuracy'].mean(),
grid.cv_results_['mean_test_roc_auc'].mean(),
grid.cv_results_['mean_test_f1'].mean())

```

```python
#"accuracy","roc_auc" and "f1" for the best combination of parameters

print('Mean accuracy:', metrics.roc_auc_score(y_test, grid_pred))
print('Mean roc_auc:', metrics.accuracy_score(y_test, grid_pred))
print('Mean f1:', metrics.f1_score(y_test, grid_pred))
```

```python
#variables with outliers and excluding correlated vars

#outlier_var=['mileage', 'hp', 'acceleration']
#non_outl_var=['model_year', 'no_of_cylinders','displacement', 'weight']

outlier_var=['hp']
non_outl_var=['model_year', 'no_of_cylinders']


cat_var=['country']
cars_data[outlier_var]
```

```python
#feature/target split

X = cars_data.drop(['car','price'], axis=1)
#nba_data[nba_data.columns[~nba_data.columns.isin(['name','target_5yrs'])]]

Y=cars_data["price"]
```

```python

#train/test split


X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#check PCA number of dimensions#
X_pca = X_train.drop(['country'], axis=1).dropna().reset_index(drop=True)

pca = PCA()
pca.fit(X_pca)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
plt.plot(cumsum)
plt.axvline(x=d, color='k', linestyle='--')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')

print("Number of components:", d)
```

```python
# create the preprocessing pipelines for both outlier and non outlier data.

cat_transformer = Pipeline(steps=[
    #('pow_transf',  PowerTransformer(method='yeo-johnson',standardize=True)), #using yeo-johnson since 'box-cox' admint only  strictly positive and there are 0 present
    #('scl', StandardScaler()), #no scaller needed since box_cox is doing the standarization
    #('pca', PCA(n_components=4))
    #('imputer', SimpleImputer(strategy='constant', fill_value='missing')), # no imputer needed since non missing values
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # no one hot needed since there is not categorical data present
    ])



outl_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean') ),
    ('pow_transf',  PowerTransformer(method='yeo-johnson',standardize=True)), #using yeo-johnson since 'box-cox' admint only  strictly positive and there are 0 present
    #('scl', StandardScaler()), #no scaller needed since box_cox is doing the standarization
    #('pca', PCA(n_components=4))
     #('onehot', OneHotEncoder(handle_unknown='ignore'))  # no one hot needed since there is not categorical data present
    ])


non_outl_transformer = Pipeline(steps=[
    #('pca', PCA(n_components=4)),
    ('imputer', SimpleImputer(strategy='mean')),
    ('scl', StandardScaler())
    #('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])


preprocessor = ColumnTransformer(
    transformers=[
        ('outliar', outl_transformer, outlier_var),
        ('nonoutliar', non_outl_transformer, non_outl_var),
        ('cat', cat_transformer, cat_var)])


# Append classifier to preprocessing pipeline.
reg = Pipeline(steps=[('preprocessor', preprocessor),
                      ('pca', PCA(n_components=2)), #adding PCA to the total of features
                      ('regressor', MLPRegressor(solver='lbfgs',max_iter=10000))])
```

```python


reg.fit(X_train, y_train)


y_pred = reg.predict(X_test)

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R Squared:', metrics.r2_score(y_test, y_pred))
```

```python
#Gridd search


parameter_space = {
    'pca__n_components': np.arange(1, 7, 2),
    'regressor__hidden_layer_sizes': [(100,), (100,50,),(120,80,40)], #[(60+5*i, ) for i in range(10)],
    'regressor__activation': [ 'tanh', 'relu'],
    'regressor__solver': ['lbfgs'],
    #'clf__alpha': 10.0 ** -np.arange(1, 10
}




grid = GridSearchCV(reg, parameter_space, n_jobs= 3, cv=5,scoring ="neg_root_mean_squared_error")
grid.fit(X_train, y_train)

print('Grid best parameter: ',grid.best_params_)
print('Grid best root_mean_squared_error: ', grid.best_score_)

grid_pred = grid.predict(X_test)
```

```python
#Gridd search


parameter_space = {
    'pca__n_components': np.arange(1, 7, 2),
    'regressor__hidden_layer_sizes': [(200,100), (100,50,)], #[(60+5*i, ) for i in range(10)],
    'regressor__activation': [ 'tanh', 'relu'],
    'regressor__solver': ['lbfgs'],
    #'clf__alpha': 10.0 ** -np.arange(1, 10
}




grid = GridSearchCV(reg, parameter_space, n_jobs= 3, cv=5,scoring ="r2")
grid.fit(X_train, y_train)

print('Grid best parameter: ',grid.best_params_)
print('Grid best R2: ', grid.best_score_)

grid_pred = grid.predict(X_test)
```
